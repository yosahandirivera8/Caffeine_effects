## Background and Prior Work


Caffeine; what a great way to start the day! Thousands of students drink caffeine to help them get through the day, especially during midterm and finals week, but how does caffeine affect the cognitive performance of younger students compared to older students? Caffeine is a natural stimulant found in various plants, however, it is mostly well-known in drinks such as coffee,  energy drinks, and teas used for a boost of energy and increased alertness. According to the FDA, the recommended caffeine intake is up to 400 milligrams per day (up to two or three 12-fluid-ounce cups). However, most people will drink copious amounts of caffeine and some even suffer withdrawal symptoms if they don’t. Some symptoms include but are not limited to fatigue, headaches, and anxiety. Relevant to this, high caffeine intake can lead to heart palpitations, dehydration, increased risk of cardiovascular issues, increased anxiety, and bone density loss. Because of this, we are concerned how caffeine intake and sleep quality affect students’ cognitive performance specifically in working memory, attention, and problem-solving abilities.

An article by Victor G. Aeby et al. highlighted the effect of energy drink consumption on college students and found that 51% of participants between the ages 21+ drank more than one energy drink each month where insufficient sleep was the most common reason. They state “Sleep deprivation is associated with selecting less difficult cognitive tasks and college students who have sleep difficulties report a greater frequency of stress” (Aeby, et al. 2007). Even though many college students rely on energy drinks to improve their alertness and treat sleep deprivation, the health consequences are alarming. Another study from Sophie Turner, et al, researched the effects of caffeine on New Zealand adolescents between the ages of 15-18. They questioned 218 participants on their daily caffeine intake and found that “the proportion of adolescents who consume caffeine daily may have risen from 71% in 2008/2009 to 94.9%” (Turner et al, 2023). They also argue that adolescents are even more vulnerable to the effects of caffeine due to their size, lack of tolerance, and a change in their mental and physical development. Many young students drink caffeine for its taste, to “look cool” or look more independent, but the fact is that drinking such high intake, especially at a young age could have detrimental effects on their bodies and could lead to several long-term health effects as mentioned above. 

Additionally, as stated above our metrics of interest–working memory, problem-solving, and attention–are crucial for evaluating the impact of caffeine on cognitive performance. Reaction time is essentially important since caffeine is a stimulant that enhances mental processing and alertness. This will help us understand and evaluate the difference in students' reaction times pre- and post-caffeine. This connects with our problem-solving metric because quick decision-making directly influences reaction time. Problem-solving relies heavily on one’s speed to process and apply information to solve problems which is tied to attention and reaction time.
Furthermore, attention is also linked to reaction time, as the ability to respond quickly requires focus and quick thinking. For example, when taking a timed exam it is recommended that students are well rested the night before since they need to be able to focus and process information at a fast pace to finish in the allotted time, however, if their reaction time is impaired, they cannot focus properly and thus their attention is affected which blocks their ability to problem-solve affecting their performance on the exam. In all, it is important to know our limits with caffeine because while it can be a great stimulant, it can lead to concerning consequences.

This difference is expected because younger individuals may be more susceptible to caffeine’s effects due to being smaller in size and they are still developing both mentally and physically which leads to an increase in sensitivity to the side effects. To see if there is a relationship between caffeine intake, sleep quality (PSQI_total), and sleep efficiency we ran a correlation matrix and a heatmap. 

The correlation matrix helps us see that there is a negative relationship between caffeine consumption and sleep efficiency. 
The heatmap reinforces the trend that higher caffeine consumption is possibly linked to worse sleep quality. This means that there is a positive correlation between the two variables. 

```python
# Import necessary libraries
import pandas as pd
from scipy.stats import ttest_ind  # Ensure ttest_ind is imported


# Define age groups
group_15_19 = df_filtered[df_filtered["age"] < 20]
group_20_25 = df_filtered[df_filtered["age"] >= 20]

# Perform independent t-tests to compare caffeine intake and sleep quality between age groups
t_stat_caffeine, p_value_caffeine = ttest_ind(group_15_19["caffeine_per_day"].dropna(), 
                                              group_20_25["caffeine_per_day"].dropna(), 
                                              equal_var=False)

t_stat_sleep_quality, p_value_sleep_quality = ttest_ind(group_15_19["PSQI_total"].dropna(), 
                                                        group_20_25["PSQI_total"].dropna(), 
                                                        equal_var=False)

t_stat_sleep_efficiency, p_value_sleep_efficiency = ttest_ind(group_15_19["Sleep_Efficiency"].dropna(), 
                                                              group_20_25["Sleep_Efficiency"].dropna(), 
                                                              equal_var=False)

# Store results in a dataframe
t_test_results = pd.DataFrame({
    "Metric": ["Caffeine Per Day", "PSQI Total (Sleep Quality)", "Sleep Efficiency"],
    "T-Statistic": [t_stat_caffeine, t_stat_sleep_quality, t_stat_sleep_efficiency],
    "P-Value": [p_value_caffeine, p_value_sleep_quality, p_value_sleep_efficiency]
})

# Display results
print("\nT-Test Results (Comparison Between Age Groups 15-19 and 20-25):\n")
t_test_results
```

To help us compare the effect that caffeine has on sleep quality between the two age groups, a t-test was conducted. Since the caffeine consumption per day had a p-value of 0.314, we can conclude that there is no statistically significant difference between the two age groups. Suggesting that the rate of caffeine consumption per day is about the same for both age groups. For the sleep quality (PSQI Total) the p-value was 0.126. This suggests that there is also no strong difference in PSQI scores between the two age groups. However, the trend may suggest that students from ages 15-19 may have slightly worse quality sleep. Finally, the p-value score for sleep efficiency was 0.056, being borderline significant. This score suggests that the sleep efficiency might be lower in the age group of 15-19 than those ages 20-25.

# Discussion and Conclusion

During this project, we focused on looking at the relationship between caffeine intake, sleep quality, and cognitive performance. Some of our variables included but not limited to were caffeine intake, age, sleep quality, and cognitive tasks. As mentioned previously, in order to accurately represent our data, we made sure to clean and only include columns we were interested in. Then, we created several barplots and scatterplots, and a t-test to analyze the data showing the relationship between our variables of interest.

We started by creating a caffeine intake plot since it is essential to understand the effect of different caffeine content in beverages on our target age groups. As expected we found that beverages such as coffee and energy drinks have a higher caffeine content. Knowing this will help us effectively measure between sleep and cognitive performance in our target age groups. Because different drinks have varying amounts of caffeine, their impact on individuals will vary. 

The Sleep Efficiency dataset and The Effects Caffeine on Sleep dataset produce comparable results since they both examine how much coffee influences sleep quality. After studying the correlation matrices for both datasets, it is clear that the results are comparable. In the Sleep efficiency statistics, we notice variations between the two age groups, with younger participants appearing to be more affected by caffeine, resulting in lower sleep quality. In contrast, older participants demonstrated that caffeine intake had little effect on their sleep quality. Similarly, The Effects of Caffeine on Sleep data demonstrate that there is a difference in sleep efficiency between the two age groups even though their PSQI scores don’t show any strong differences. 

To investigate further into whether the effects of caffeine could result in consequences in cognitive performance, we utilized the dataset: Sleep Deprivation and Cognitive Performance. The data examined in this dataset contained factors such as sleep quality alongside caffeine intake and the results from cognitive performance tasks between our two specific age groups. It was shown that participants who were 20-25 consumed more caffeine than those who were 18-19. However, caffeine consumption only affected the younger participants during the Stroop Task and there was no significant difference in their performance during the accuracy and stimuli tasks. In comparison to the older participants, there was an opposite effect with higher caffeine intake leading to faster reaction times in the Stroop task and higher accuracy rates in the working memory task. But, similarly to the younger group, there was not enough sufficient evidence to demonstrate that the visual stimulus task was affected. From these results, it is suggested that perhaps older individuals who have grown accustomed to increased caffeine consumption are able to withstand the negative effects of caffeine and enhance their performance in cognitive tasks better than those who are younger. Additionally, the sleep quality data presented was able to verify that for the 18-19 group, better sleep quality improved their cognitive performance more than those part of the 20-25 group. Based on these findings, we can infer that younger participants are more reliant on adequate sleep for optimal cognitive performance while older participants gain cognitive benefits from caffeine.

One of the limitations we faced was that three out of the four datasets we used did not directly measure cognitive performance tasks like memory tests, reaction times, or problem-solving tasks. Instead, two of our sleeping datasets inferred a cognitive impact based on sleep efficiency and PSQI scores, which may not fully capture the students' cognitive ability. However, despite not having direct access to datasets that had information on cognitive performance tasks, we found a few research articles that we will use to cross-reference our analysis and help us answer our research question. Studies have shown that caffeine intake can enhance explicit memory performance, particularly in young adults,  especially when consumed during non-optimal times of day, for example in the mornings, when our cognitive performance is naturally lower (Sherman et al., 2016). This finding suggests that students who consume caffeine to compensate for poor sleep might experience a short-term boost in memory performance. However, in a study done by Soos et al., they suggest that long-term reliance on caffeine as a compensatory mechanism for poor sleep efficiency may not be beneficial and eventually disrupt their sleep efficiency. Additionally, research highlights that excessive caffeine consumption, especially through energy drinks, is associated with negative cognitive performance in adolescents and young adults. Soos et al. have linked high caffeine intake to increased stress, anxiety, and even impaired sleeping patterns, all of which are known to hinder cognitive performance (Soos et al., 2021). This aligns with our findings from the Sleep Efficiency dataset and the Effects of Caffeine on Sleep dataset, as they suggest that the higher the caffeine consumption the worse the quality of sleep in adolescents is. By integrating these research findings with our dataset analysis, we provide stronger evidence to help answer our research question.

For our study, we decided to focus on how caffeine intake and sleep quality influence cognitive performance—specifically working memory, attention, and problem-solving abilities—among students aged 15-19 compared to those aged 20-25. Our findings indicate that caffeine affects cognitive performance differently depending on age, with younger students appearing more vulnerable to its negative effects, while older students benefit from it in certain cognitive tasks. Younger students (18-19) experienced more disruptions in sleep quality due to caffeine intake, whereas older students’ sleep efficiency was less affected, suggesting greater caffeine tolerance in the older group. Additionally, caffeine negatively impacted attention (Stroop Task) in younger students, leading to slower reaction times, while working memory and sustained attention were not significantly affected. In contrast, older students (20-25) demonstrated improved working memory accuracy (N-Back Task) and faster reaction times on attention tasks when consuming caffeine, while sustained attention remained unchanged. Interestingly, better sleep quality was associated with improved cognitive performance in younger students, whereas older students appeared to rely less on sleep and more on caffeine to enhance cognitive function.

Our hypothesis was partially correct as caffeine intake significantly affected cognitive performance across age groups but in different ways. While we correctly predicted that younger individuals would be more sensitive to caffeine's negative effects, we did not anticipate that older students would experience cognitive benefits, particularly in working memory tasks. Moreover, sleep quality was found to moderate cognitive performance for younger students, but its impact on older students was less pronounced. These findings suggest that younger individuals are more reliant on adequate sleep for optimal cognitive performance, whereas older individuals may have adapted to caffeine and use it to compensate for sleep deficits. While caffeine impaired attention in younger students, it appeared to enhance cognitive processing speed and working memory in older students, indicating that habitual caffeine consumption and tolerance could play a role in mitigating its negative effects in older individuals.

To strengthen these findings, future research should analyze long-term caffeine tolerance in different age groups, assess the role of stress and academic pressure as potential confounding variables, incorporate objective sleep-tracking measures, and examine caffeine withdrawal effects to better understand how habitual use influences cognitive function over time. Understanding the interaction between caffeine, sleep, and cognitive performance is crucial for students, educators, and health professionals, as it provides insight into how caffeine consumption and sleep efficiency impact cognitive abilities across different age groups.
